Restaurant Menu Display

```python
menu = {
    "Veg": {
        "Veg Biryani": 120,
        "Paneer Butter Masala": 150,
        "Butter Naan": 30
    },
    "Non-Veg": {
        "Chicken Biryani": 180,
        "Egg Curry": 100
    },
    "Hot Drinks": {
        "Tea": 15,
        "Coffee": 20
    },
    "Cold Drinks": {
        "Cold Drink": 40,
        "Lassi": 35
    },
    "Desserts": {
        "Gulab Jamun": 50,
        "Ice Cream": 60
    },
    "Breakfast": {
        "Idli": 30,
        "Dosa": 40
    },
    "Lunch": {
        "Meals": 100,
        "Curd Rice": 70
    },
    "Dinner": {
        "Chapathi": 50,
        "Parotta": 45
    }
}

order = {}
total = 0

print("=== Welcome to Our Restaurant ===")

while True:
    print("\nCategories:")
    for idx, category in enumerate(menu.keys(), 1):
        print(f"{idx}. {category}")
    print("Type 'done' to finish ordering.")

    choice = input("Choose a category: ").strip()
    if choice.lower() == 'done':
        break

    matched_category = None
    for cat in menu:
        if cat.lower() == choice.lower():
            matched_category = cat
            break

    if matched_category:
        print(f"\n--- {matched_category} Menu ---")
        for item, price in menu[matched_category].items():
            print(f"{item}: ₹{price}")
        
        while True:
            item_name = input("Enter item to order (or type 'back' to choose another category): ").strip()
            if item_name.lower() == 'back':
                break
            if item_name in menu[matched_category]:
                qty = int(input(f"Enter quantity for {item_name}: "))
                order[item_name] = order.get(item_name, 0) + qty
                total += menu[matched_category][item_name] * qty
            else:
                print("Item not found in this category.")
    else:
        print("Invalid category. Please try again.")


print("\n=== Your Bill ===")
for item, qty in order.items():
    # Find the item's price
    for cat in menu:
        if item in menu[cat]:
            price = menu[cat][item]
            print(f"{item} x{qty} = ₹{price * qty}")
print(f"Total Amount: ₹{total}")
print("Thank you for visiting!")
```
